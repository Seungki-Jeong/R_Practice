```r
library(rvest)
library(stringr)
library(dplyr)
```

```r
title = c()
body = c()
```

```r
url_1 = "https://www.bskorea.or.kr/bible/korbibReadpage.php?version=GAE&book=gen&chap="
url_2 = "&sec=1&cVersion=&fontSize=15px&fontWeight=normal"
```

```r
url = paste(url_1, 1, url_2, sep = "")
```

```r
title_css = ".smallTitle"
body_css = "#tdBible1 span"
```

```r
hdoc = read_html(url, encoding = 'utf-8')
```

```r
t_css = html_nodes(hdoc, title_css)
t_part = html_text(t_css)
t_part
```

```r
b_css = html_nodes(hdoc, body_css)
b_part = html_text(b_css)
```

```r
body_part = gsub('\\d+', "", b_part)
```

```r
body_part_1 = str_trim(body_part, side = 'both')
```

```r
body_ = c(body, body_part_1)
```

```r
for(i in 1:50){
    url = paste(url_1, i, url_2, sep = "")
    title_css = ".smallTitle"
    body_css = "#tdBible1 span"
    hdoc = read_html(url, encoding = 'utf-8')
    t_css = html_nodes(hdoc, title_css)
    t_part = html_text(t_css)
    b_css = html_nodes(hdoc, body_css)
    b_part = html_text(b_css)
    body_part = gsub('\\d+', "", b_part)
    body_part_1 = str_trim(body_part, side = 'both')
    title_ = c(title, t_part)
    body_ = c(body, body_part_1)
}
```
